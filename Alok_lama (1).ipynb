Problem - 1: Getting Started with Data Exploration

Data exploring and understanding

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

world_happiness=pd.read_csv('/content/drive/MyDrive/Copy of WHR-2024-5CS037.csv')

#DATASET OVERVIEW
print("The first 10 rows are:")
print(world_happiness.head(10))

rows, columns = world_happiness.shape
print("rows:", rows)
print("columns", columns)

print("data types:\n", world_happiness.dtypes)

#BASIC STATISTICS
Country_mean= world_happiness['score'].mean()
Country_median=world_happiness['score'].median()
Country_standard_deviation=world_happiness['score'].std()

print("\n Mean:", Country_mean)
print("Median:", Country_median)
print("Standard Deviation:", Country_standard_deviation)


highest=world_happiness.nlargest(1,'score')[ 'Country name'].values[0]
lowest=world_happiness.nsmallest(1, 'score')[ 'Country name'].values[0]

print("\nHighest Score:", highest)
print("Lowest Score:", lowest)

#MISSING VALUES
missing = world_happiness.isnull().sum()
print(" Total missing values per coulumn:")
print(missing)

#FILTERING AND SORTING
filtered = world_happiness[world_happiness['score']>7.5]
print("Countries with score greater than 7.5 are:")
print(filtered['Country name'])

filtered_top10 = filtered.sort_values( by = 'Log GDP per capita', ascending = False).head(10)
print(filtered_top10['Country name'])
print("The top 10 rows contain obly 3 countires as only three countries have score greater than  7.5.")

#Adding New columns
conditions = [ (world_happiness['score'] <4),( world_happiness['score']>=4 ), (world_happiness['score']> 6)]
categories = [ 'Low', 'Medium', 'High']
world_happiness['Happiness Category'] = np.select(conditions, categories)
print(world_happiness[['Country name', 'score','Happiness Category']])

#Data visulaization

#line
top_10_happiest =world_happiness.nlargest(10, 'score')
plt.figure(figsize=(10, 6))
sns.barplot(x='Country name', y='score', data=top_10_happiest)
plt.title('Top 10 Happiest Countries by Score')
plt.xlabel('Country name')
plt.ylabel('Happiness Score')
plt.show()
plt.figure(figsize=(10, 6))


# Line Plot: Top 10 unhappiest countries by Score
top_10_unhappiest = world_happiness.nsmallest(10, 'score')
plt.figure(figsize=(10, 6))
plt.plot(top_10_unhappiest['Country name'], top_10_unhappiest['score'], marker='o', color='red', linestyle='-')
plt.title('Top 10 Unhappiest Countries by Score')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Histogram: Distribution of the Score column
plt.figure(figsize=(8, 5))
sns.histplot(world_happiness['score'], kde=True, bins=20, color='red')
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Scatter Plot: GDP per Capita vs. Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=world_happiness, hue='Happiness Category', palette='cool', s=100)
plt.title('Relationship between GDP per Capita and Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.tight_layout()
plt.show()
```

Problem - 2 - Some Advance Data Exploration Task

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

world_happiness = pd.read_csv('/content/drive/MyDrive/Copy of WHR-2024-5CS037.csv')

# Preparing south-asia Dataset
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                         "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

filtered_southasia = world_happiness[world_happiness['Country name'].isin(south_asian_countries)]

filtered_southasia.to_csv('filtered_data.csv', index=False)
print(filtered_southasia)

# composite score ranking
filtered_southasia['Composite Score'] = (0.40 * filtered_southasia['Log GDP per capita'] +
                                         0.30 * filtered_southasia['Social support'] +
                                         0.30 * filtered_southasia['Healthy life expectancy'])

rank_southasia = filtered_southasia.sort_values(by='Composite Score', ascending=False)
print(rank_southasia[['Country name', 'Composite Score']])
plt.figure(figsize=(10, 6))
sns.barplot(
    data=rank_southasia,
    x='Country name', y='Composite Score', palette='cool'
)
plt.title('Ranking of South Asian Countries by Composite Score', fontsize=16)
plt.xlabel('Country name')
plt.ylabel('Composite Score')
plt.show()

#comparing composite score with original score
plt.figure(figsize=(10, 6))

# Scatter plot comparing Composite Score and Happiness Score
sns.scatterplot(
    data=filtered_southasia, x='Composite Score', y='score',
    hue='Country name', palette='tab10', s=100
)
plt.title('Composite Score vs Happiness Score', fontsize=16)
plt.xlabel('Composite Score')
plt.ylabel('Happiness Score')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# Bar chart comparison of Composite Score and Happiness Score
filtered_southasia[['Country name', 'Composite Score', 'score']].set_index('Country name').plot(
    kind='bar', figsize=(10, 6), colormap='coolwarm'
)
plt.title('Comparison of Composite and Happiness Scores', fontsize=16)
plt.ylabel('Scores')
plt.tight_layout()
plt.show()

#Task-3-Outlier Direction

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=filtered_southasia, x='Log GDP per capita', y='score',
    hue='Country name', palette='tab10', s=100
)
plt.title('Happiness Score vs Log GDP per Capita (South Asia)', fontsize=16)
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

filtered_southasia['GDP Z-Score'] = zscore(filtered_southasia['Log GDP per capita'])
filtered_southasia['Happiness Z-Score'] = zscore(filtered_southasia['score'])

outliers = filtered_southasia[
    (filtered_southasia['GDP Z-Score'].abs() > 2) |
    (filtered_southasia['Happiness Z-Score'].abs() > 2)
]

print("Outlier Countries:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])

# Define outliers using the 1.5 × IQR rule.
Q1_gdp = filtered_southasia['Log GDP per capita'].quantile(0.25)
Q3_gdp = filtered_southasia['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

Q1_happiness = filtered_southasia['score'].quantile(0.25)
Q3_happiness = filtered_southasia['score'].quantile(0.75)
IQR_happiness = Q3_happiness - Q1_happiness

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp
lower_bound_happiness = Q1_happiness - 1.5 * IQR_happiness
upper_bound_happiness = Q3_happiness + 1.5 * IQR_happiness

outliers = filtered_southasia[
    (filtered_southasia['Log GDP per capita'] < lower_bound_gdp) |
    (filtered_southasia['Log GDP per capita'] > upper_bound_gdp) |
    (filtered_southasia['score'] < lower_bound_happiness) |
    (filtered_southasia['score'] > upper_bound_happiness)
]

print("Outlier Countries:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])




#a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
#in a different color
filtered_southasia['Outlier'] = filtered_southasia.apply(
    lambda row: 'Outlier' if
    (row['Log GDP per capita'] < lower_bound_gdp or row['Log GDP per capita'] > upper_bound_gdp or
     row['score'] < lower_bound_happiness or row['score'] > upper_bound_happiness)
    else 'Non-Outlier', axis=1
)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=filtered_southasia, x='Log GDP per capita', y='score',
    hue='Outlier', palette={'Outlier': 'red', 'Non-Outlier': 'blue'}, s=100
)
for _, row in filtered_southasia.iterrows():
    plt.text(row['Log GDP per capita'], row['score'], row['Country name'], fontsize=8)

plt.title('Score vs Log GDP per Capita (Outliers Highlighted)', fontsize=16)
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.grid(True)
plt.legend(title='Legend')
plt.show()


print("""
 Countries with high GDP but low happiness may suffer from inequality, political instability, or lack of social support.
 Countries with low GDP but high happiness may have strong social bonds, cultural values, or community support that enhance well-being
 Some countries show extreme values, influenced by unique social, economic, or environmental factors.
"""
)


#Exploring trends across metrics
# Calculate Pearson correlation between Freedom to Make Life Choices, Generosity, and Score
correlation_freedom = filtered_southasia[['Freedom to make life choices', 'score']].corr(method='pearson')
correlation_generosity = filtered_southasia[['Generosity', 'score']].corr(method='pearson')

# Display the correlation results
print("Correlation between Freedom to Make Life Choices and Score:")
print(correlation_freedom)

print("\nCorrelation between Generosity and Score:")
print(correlation_generosity)



# Scatter plot with trendline for Freedom to Make Life Choices vs Score
plt.figure(figsize=(10, 6))
sns.regplot(
    data=filtered_southasia, x='Freedom to make life choices', y='score',
    scatter_kws={'s': 100, 'color': 'blue'}, line_kws={'color': 'red'}
)
plt.title('Freedom to Make Life Choices vs Score', fontsize=16)
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Score')
plt.grid(True)
plt.show()

# Scatter plot with trendline for Generosity vs Score
plt.figure(figsize=(10, 6))
sns.regplot(
    data=filtered_southasia, x='Generosity', y='score',
    scatter_kws={'s': 100, 'color': 'green'}, line_kws={'color': 'orange'}
)
plt.title('Generosity vs Score', fontsize=16)
plt.xlabel('Generosity')
plt.ylabel('Score')
plt.grid(True)
plt.show()

print("""
Freedom to Make Life Choices vs. Score:
High correlation: Countries with more personal freedom tend to have higher happiness scores. Greater autonomy in life decisions leads to better well-being.

Generosity vs. Score:
Low correlation: Generosity doesn't strongly impact happiness scores. Other factors like economic stability and social support are likely more influential in South Asia.
""")

#Task 5: Gap Analysis
filtered_southasia['GDP-Score Gap'] = filtered_southasia['Log GDP per capita'] - filtered_southasia['score']

print(filtered_southasia[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])



rank_ascending = filtered_southasia.sort_values(by='GDP-Score Gap', ascending=True)
print("Ranking by GDP-Score Gap (Ascending):")
print(rank_ascending[['Country name', 'GDP-Score Gap']])

rank_descending = filtered_southasia.sort_values(by='GDP-Score Gap', ascending=False)
print("\nRanking by GDP-Score Gap (Descending):")
print(rank_descending[['Country name', 'GDP-Score Gap']])

top_positive_gap = rank_descending.head(3)
top_negative_gap = rank_ascending.head(3)

highlighted_gaps = pd.concat([top_positive_gap, top_negative_gap])

plt.figure(figsize=(10, 6))
sns.barplot(
    data=highlighted_gaps, x='Country name', y='GDP-Score Gap', palette='coolwarm'
)
plt.title('Top 3 Countries with Largest Positive and Negative GDP-Score Gaps', fontsize=16)
plt.xlabel('Country Name')
plt.ylabel('GDP-Score Gap')
plt.grid(True)
plt.show()


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

world_happiness=pd.read_csv('/content/drive/MyDrive/Copy of WHR-2024-5CS037.csv')

#Descriptive statistics
middle_east_countries = [ "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
"Kuwait", "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia", "Syria",
"United Arab Emirates", "Yemen"]



filtered_middle_east = world_happiness[world_happiness['Country name'].isin(middle_east_countries)]

south_asia_stats = filtered_southasia['score'].describe()[['mean', 'std']]
middle_east_stats = filtered_middle_east['score'].describe()[['mean', 'std']]

print("South Asia - Mean and Standard Deviation of Score:")
print(south_asia_stats)

print("\nMiddle East - Mean and Standard Deviation of Score:")
print(middle_east_stats)

if south_asia_stats['mean'] > middle_east_stats['mean']:
    print("\nSouth Asia has a higher average happiness score.")
else:
    print("\nMiddle East has a higher average happiness score.")

#Top and Bottom Performers

top_3_southasia = filtered_southasia.nlargest(3, 'score')
bottom_3_southasia = filtered_southasia.nsmallest(3, 'score')

top_3_middle_east = filtered_middle_east.nlargest(3, 'score')
bottom_3_middle_east = filtered_middle_east.nsmallest(3, 'score')

top_southasia = top_3_southasia[['Country name', 'score']]
bottom_southasia = bottom_3_southasia[['Country name', 'score']]
top_middle_east = top_3_middle_east[['Country name', 'score']]
bottom_middle_east = bottom_3_middle_east[['Country name', 'score']]

combined_data = pd.concat([top_southasia, bottom_southasia, top_middle_east, bottom_middle_east])

plt.figure(figsize=(12, 6))
sns.barplot(data=combined_data, x='Country name', y='score', hue='Country name', palette='Set2')
plt.title('Top 3 and Bottom 3 Countries by Happiness Score (South Asia vs. Middle East)', fontsize=16)
plt.xlabel('Country Name')
plt.ylabel('score')
plt.xticks(rotation=45)
plt.show()


#Metric Comparisions
south_asia_metrics = filtered_southasia[['Country name', 'Log GDP per capita', 'Social support', 'Healthy life expectancy']]
middle_east_metrics = filtered_middle_east[['Country name', 'Log GDP per capita', 'Social support', 'Healthy life expectancy']]

south_asia_metrics['Region'] = 'South Asia'
middle_east_metrics['Region'] = 'Middle East'

combined_metrics = pd.concat([south_asia_metrics, middle_east_metrics])

melted_data = pd.melt(combined_metrics, id_vars=['Country name', 'Region'], value_vars=['Log GDP per capita', 'Social support', 'Healthy life expectancy'],
                      var_name='Metric', value_name='Value')

plt.figure(figsize=(14, 7))
sns.barplot(data=melted_data, x='Country name', y='Value', hue='Region', palette='Set2')
plt.title('Comparison of Key Metrics Between South Asia and Middle East', fontsize=16)
plt.xlabel('Country Name')
plt.ylabel('Metric Value')
plt.xticks(rotation=45)
plt.show()

print("GDP per capita and Health Life expectancy show the largest disparity.")


#Happiness Disparity
south_asia_range = filtered_southasia['score'].max() - filtered_southasia['score'].min()
middle_east_range = filtered_middle_east['score'].max() - filtered_middle_east['score'].min()

south_asia_cv = filtered_southasia['score'].std() / filtered_southasia['score'].mean()
middle_east_cv = filtered_middle_east['score'].std() / filtered_middle_east['score'].mean()

print(f"South Asia - Range: {south_asia_range}, Coefficient of Variation: {south_asia_cv}")
print(f"Middle East - Range: {middle_east_range}, Coefficient of Variation: {middle_east_cv}")

print("South Asia has greater variablity in happinesss score.")


#Correlation Analysis
south_asia_corr_freedom = filtered_southasia['score'].corr(filtered_southasia['Freedom to make life choices'])
south_asia_corr_generosity = filtered_southasia['score'].corr(filtered_southasia['Generosity'])

middle_east_corr_freedom = filtered_middle_east['score'].corr(filtered_middle_east['Freedom to make life choices'])
middle_east_corr_generosity = filtered_middle_east['score'].corr(filtered_middle_east['Generosity'])

print(f"South Asia - Correlation with Freedom to Make Life Choices: {south_asia_corr_freedom}")
print(f"South Asia - Correlation with Generosity: {south_asia_corr_generosity}")
print(f"Middle East - Correlation with Freedom to Make Life Choices: {middle_east_corr_freedom}")
print(f"Middle East - Correlation with Generosity: {middle_east_corr_generosity}")

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=filtered_southasia, x='Freedom to make life choices', y='score', color='blue')
plt.title('South Asia: Score vs Freedom to Make Life Choices')

plt.subplot(1, 2, 2)
sns.scatterplot(data=filtered_southasia, x='Generosity', y='score', color='green')
plt.title('South Asia: Score vs Generosity')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=filtered_middle_east, x='Freedom to make life choices', y='score', color='blue')
plt.title('Middle East: Score vs Freedom to Make Life Choices')

plt.subplot(1, 2, 2)
sns.scatterplot(data=filtered_middle_east, x='Generosity', y='score', color='green')
plt.title('Middle East: Score vs Generosity')

plt.tight_layout()
plt.show()

#Outlier Detection
south_asia_score_q1 = filtered_southasia['score'].quantile(0.25)
south_asia_score_q3 = filtered_southasia['score'].quantile(0.75)
south_asia_score_iqr = south_asia_score_q3 - south_asia_score_q1
south_asia_score_lower_bound = south_asia_score_q1 - 1.5 * south_asia_score_iqr
south_asia_score_upper_bound = south_asia_score_q3 + 1.5 * south_asia_score_iqr

south_asia_gdp_q1 = filtered_southasia['Log GDP per capita'].quantile(0.25)
south_asia_gdp_q3 = filtered_southasia['Log GDP per capita'].quantile(0.75)
south_asia_gdp_iqr = south_asia_gdp_q3 - south_asia_gdp_q1
south_asia_gdp_lower_bound = south_asia_gdp_q1 - 1.5 * south_asia_gdp_iqr
south_asia_gdp_upper_bound = south_asia_gdp_q3 + 1.5 * south_asia_gdp_iqr

south_asia_score_outliers = filtered_southasia[(filtered_southasia['score'] < south_asia_score_lower_bound) |
                                              (filtered_southasia['score'] > south_asia_score_upper_bound)]

south_asia_gdp_outliers = filtered_southasia[(filtered_southasia['Log GDP per capita'] < south_asia_gdp_lower_bound) |
                                              (filtered_southasia['Log GDP per capita'] > south_asia_gdp_upper_bound)]

middle_east_score_q1 = filtered_middle_east['score'].quantile(0.25)
middle_east_score_q3 = filtered_middle_east['score'].quantile(0.75)
middle_east_score_iqr = middle_east_score_q3 - middle_east_score_q1
middle_east_score_lower_bound = middle_east_score_q1 - 1.5 * middle_east_score_iqr
middle_east_score_upper_bound = middle_east_score_q3 + 1.5 * middle_east_score_iqr

middle_east_gdp_q1 = filtered_middle_east['Log GDP per capita'].quantile(0.25)
middle_east_gdp_q3 = filtered_middle_east['Log GDP per capita'].quantile(0.75)
middle_east_gdp_iqr = middle_east_gdp_q3 - middle_east_gdp_q1
middle_east_gdp_lower_bound = middle_east_gdp_q1 - 1.5 * middle_east_gdp_iqr
middle_east_gdp_upper_bound = middle_east_gdp_q3 + 1.5 * middle_east_gdp_iqr

middle_east_score_outliers = filtered_middle_east[(filtered_middle_east['score'] < middle_east_score_lower_bound) |
                                                  (filtered_middle_east['score'] > middle_east_score_upper_bound)]

middle_east_gdp_outliers = filtered_middle_east[(filtered_middle_east['Log GDP per capita'] < middle_east_gdp_lower_bound) |
                                                  (filtered_middle_east['Log GDP per capita'] > middle_east_gdp_upper_bound)]

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=filtered_southasia, x='Log GDP per capita', y='score', color='blue', label='South Asia')
sns.scatterplot(data=south_asia_score_outliers, x='Log GDP per capita', y='score', color='red', label='Score Outliers')
sns.scatterplot(data=south_asia_gdp_outliers, x='Log GDP per capita', y='score', color='green', label='GDP Outliers')
plt.title('South Asia Outliers (Score and GDP per Capita)')

plt.subplot(1, 2, 2)
sns.scatterplot(data=filtered_middle_east, x='Log GDP per capita', y='score', color='blue', label='Middle East')
sns.scatterplot(data=middle_east_score_outliers, x='Log GDP per capita', y='score', color='red', label='Score Outliers')
sns.scatterplot(data=middle_east_gdp_outliers, x='Log GDP per capita', y='score', color='green', label='GDP Outliers')
plt.title('Middle East Outliers (Score and GDP per Capita)')

plt.tight_layout()
plt.show()

print(f"South Asia Score Outliers:\n{south_asia_score_outliers[['Country name', 'score']]}")
print(f"South Asia GDP Outliers:\n{south_asia_gdp_outliers[['Country name', 'Log GDP per capita']]}")
print(f"Middle East Score Outliers:\n{middle_east_score_outliers[['Country name', 'score']]}")
print(f"Middle East GDP Outliers:\n{middle_east_gdp_outliers[['Country name', 'Log GDP per capita']]}")

#Visualization


filtered_southasia['Region'] = 'South Asia'
filtered_middle_east['Region'] = 'Middle East'

combined_region_data = pd.concat([filtered_southasia[['Country name', 'score', 'Region']],
                                  filtered_middle_east[['Country name', 'score', 'Region']]])

plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_region_data, x='Region', y='score', palette='Set2')
plt.title('Happiness Score Distribution: South Asia vs. Middle East', fontsize=16)
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.show()

```
